```python
# Alisha Rhodes
# 11/17/2025
# 9.2
# defining a generator function that returns the third odd number within a defined range

def get_odds():
    for number in range(10):   
        if number % 2 == 1:
            yield number

count = 0
for odd in get_odds():
    count += 1
    if count == 3:
        print("The third odd number is:", odd)
        break
```
