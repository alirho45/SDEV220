```python
# Alisha Rhodes
# Assignment 4.2
# Matching attributes
small = True
green = True

if small and green:
    print("pea")
elif small and not green:
    print("cherry")
elif not small and green:
    print("watermelon")
else:
    print("pumpkin")
```

```python
# Alisha Rhodes
# Assignment 6.2
# While loop comparing a number with a guess
guess_me = 7
number = 1

while True:
    if number < guess_me:
        print("too low")
    elif number == guess_me:
        print("found it!")
        break
    else:
        print("oops")
        break
    number += 1
```

```python
# Alisha Rhodes
# Assignment 6.3
# Determine if a number is within range
guess_me = 5

for number in range(10):
    if number < guess_me:
        print("too low")
    elif number == guess_me:
        print("found it!")
        break
    else:
        print("oops")
        break
```
