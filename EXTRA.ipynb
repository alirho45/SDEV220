```python
# Alisha Rhodes
# 12/8/2025
# EXTRA CREDIT Math Puzzle Can you find the number?

import time

# Reverse number mathematically
def rev_number(n: int):
    rev = 0
    while n > 0:
        rev = rev * 10 + n % 10
        n //= 10
    return rev

# Check using math reversal
def check_multiples(n: int, multiplier: int):
    product = n * multiplier
    return product == rev_number(n)

# General search
def find_rev_multiples(digits: int, multiplier: int):
    start = 10**(digits-1)
    end = 10**digits - 1
    
    for n in range(start, end+1):
        product = n * multiplier
        # Include any shortcuts such as recognizing that 2178 * 4 must be > 8000
        if len(str(product)) != digits:
            continue
        if check_multiples(n, multiplier):
            return n, product
    return None

#Runtime measurement
total_start = time.time()

# Examples
for digits, multiplier in [(4,4), (5,4), (5,9), (6,8)]:
    start = time.time()
    result = find_rev_multiples(digits, multiplier)
    end = time.time()
    print(f"{digits}-digit ×{multiplier}: {result}, runtime: {end - start:.4f} seconds")

total_end = time.time()
print(f"\nTotal runtime: {total_end - total_start:.4f} seconds")


```
